```python
import random
import tkinter as tk
from tkinter import messagebox

GOAL = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

buttons = []
#Tính value bằng số ô giống goal
def Value(state):
    score = 0
    for r in range(3):
        for c in range(3):
            if state[r][c] != 0 and state[r][c] == GOAL[r][c]:
                score += 1
    return score

def generate_neighbors(state):
    neighbors = []
    r0, c0 = 0, 0
    for r in range(3):
        for c in range(3):
            if state[r][c] == 0:
                r0, c0 = r, c
                break

    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for dr, dc in directions:
        new_r, new_c = r0 + dr, c0 + dc
        if 0 <= new_r < 3 and 0 <= new_c < 3:
            next_state = [row[:] for row in state]
            next_state[r0][c0], next_state[new_r][new_c] = next_state[new_r][new_c], next_state[r0][c0]
            neighbors.append(next_state)
    return neighbors

def Simple_Hill_Climbing(Start):
    Current_State = [row[:] for row in Start]
    while True:
        update_ui(Current_State)
        root.update()
        root.after(600)
        
        neighbors_list = generate_neighbors(Current_State)
        better_state_found = False
        
        for Next_State in neighbors_list:
            if Value(Next_State) > Value(Current_State):
                Current_State = Next_State
                better_state_found = True
                break
        
        if not better_state_found:
            break
            
    return Current_State

def update_ui(state):
    global current_global_state
    current_global_state = [row[:] for row in state]
    for r in range(3):
        for c in range(3):
            val = state[r][c]
            btn_index = r * 3 + c
            if val == 0:
                buttons[btn_index].config(text="", bg="#ccc")
            else:
                buttons[btn_index].config(text=str(val), bg="#fff")
    lbl_status.config(text=f"Value(Current_State) = {Value(state)} / 8")

def action_random_start():
    state = [row[:] for row in GOAL]
    num_shuffles = random.randint(5, 8)
    for _ in range(num_shuffles):
        neighbors = generate_neighbors(state)
        state = random.choice(neighbors)
    update_ui(state)

def action_solve():
    final_state = Simple_Hill_Climbing(current_global_state)
    if final_state == GOAL:
        messagebox.showinfo("Success", "Thuật toán đã tìm được trạng thái ĐÍCH thành công!")
    else:
        messagebox.showwarning("Local Maximum", "Thuật toán đã dừng lại do bị kẹt ở CỰC ĐẠI CỤC BỘ!")

root = tk.Tk()
root.title("Simple Hill Climbing (8-Puzzle)")

matrix_frame = tk.Frame(root)
matrix_frame.pack(pady=15)

for i in range(9):
    btn = tk.Button(matrix_frame, font=("Arial", 22, "bold"), width=4, height=2)
    btn.grid(row=i//3, column=i%3, padx=4, pady=4)
    buttons.append(btn)

lbl_status = tk.Label(root, font=("Arial", 12, "bold"), fg="#1d3557")
lbl_status.pack(pady=5)

control_frame = tk.Frame(root)
control_frame.pack(pady=10)

tk.Button(control_frame, text="Random Start", font=("Arial", 10), command=action_random_start).grid(row=0, column=0, padx=15)
tk.Button(control_frame, text="Run Algorithm", font=("Arial", 10, "bold"), command=action_solve, bg="#2a9d8f", fg="white").grid(row=0, column=1, padx=15)

action_random_start()
root.mainloop()
```
